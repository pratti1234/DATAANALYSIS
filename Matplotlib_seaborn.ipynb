```python
a = np.array([25,60,5,10])
labe = ["AIML","Python","Pandas","Numpy"]
c=['yellow','blue','green','red']
plt.pie(a,labels = labe,colors=c)#to get pie chart we use pie  #colours we can change by creating another list
plt.show()
```

```python
#explode
a = np.array([25,60,5,10])
labe = ["AIML","Python","Pandas","Numpy"]
explo = [0.5,0,0,0]#0.5 to increase distance
plt.pie(a,labels = labe,explode = explo,startangle = 180)#start angle to change rotation #cutting of particular pie chart value is explode function
plt.show()
```

```python
a = np.array([25,60,5,10])
labe = ["AIML","Python","Pandas","Numpy"]
explo = [0.3,0,0,0]
plt.pie(a,labels = labe,explode = explo,shadow=True)#labels for showing small box or legend function #to come shadow
plt.legend()
plt.show()
```

```python
import seaborn as sns#14 to 15 datasets are present as default
import matplotlib.pyplot as plt
#Load example dataset
iris=sns.load_dataset("iris")
print(iris)
#create a scatter plot
sns.scatterplot(x="sepal_length",y="sepal_width",data=iris,label="india")#label is to add

plt.title("scatter plot of sepal_length vs sepal_width ")
plt.xlabel("sepal_length($)")
plt.ylabel("sepal_width ($)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
titanic=sns.load_dataset("titanic")
g=sns.FacetGrid(titanic,col="class")
g.map(sns.histplot,"age")
plt.subplots_adjust(top=0.8)
g.fig.suptitle("age distribution by passenger class")
plt.show()
```
